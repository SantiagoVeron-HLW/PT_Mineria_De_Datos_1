```python
df.describe()
```

La función df.describe() muestra el resumen estadístico de las columnas numéricas, donde se pueden identificar anomalías, errores lógicos y valores atípicos (outliers) en los datos que deben ser corregidos antes de realizar cualquier análisis. Por ejemplo, en la columna 'age' el valor mínimo es -5 y el valor máximo es 150, lo que es biologicamente imposible y generaría fallas en el análisis. En la columna 'montthly_watch_time' (tiempo de reproducción mensual en minutos) se registra un valor mínimo negativo de -120 y un máximo de 99999 lo que supera a los minutos máximos de un mes. En la columna last_login_date el valor máximo es 2029, lo que no se corresponde al tiempo real ya que estamos actualmente en el año 2026. En la columna 'customer_support_tickets' el valor mínimo es de -1 y el máximo de 150, lo cual representa resultados anómalos ya que, por una parte, no se puede abrir -1 tickets y, por otra parte, 75% de los usuarios tiene un máximo de 1 ticket, representando el 150 un outlier (valor atípico) extremo.

```python
df.duplicated().sum()
```

Con esta función, podemos identificar la cantidad de registros duplicados o clonados que presenta el detaset.

```python
for columna in df.columns:
  print (columna)
  print (df[columna].unique())
  print ("-"*40)
```

La exploración de valores únicos nos muestra que, además de los valores atípicos (outliers), el dataset presenta una falta de estandarización en las columnas categóricas. Por ejemplo, en la columna 'suscription_plan' la misma variable fue introducida de diferente manera incluyendo errores como mezcla de mayusculas y minusculas, abreviaciones, errores de escritura y espacios vacios al final de las palabras. En la columna 'country' también se observa la duplicación de categorías por la falta de un formato único. Por último, en la columna 'favorite_genre' se observan los errores anteriores como la mezcla de mayuscula y minuscula; una mezcla entre idiomas y la presencia de valores nulos con el texto 'None'. Asimismo, se confirman los valores atípicos obtenidos en el 'df.describe' y se identifican otros. Por ejemplo en la columna 'age' además de los outliers -5 y 150, aparece un valor de 130, lo que también es biologicamente imposible. Debido a esto, es necesario aplicar técnicas de estandarización para obtener un dataset listo para análisis.
